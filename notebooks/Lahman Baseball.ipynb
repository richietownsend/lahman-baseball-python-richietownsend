### Load Tables from PostgreSQL

```python
college_df = pd.read_sql("SELECT * FROM collegeplaying", con=engine)
# college_df.head()
```

```python
people_df = pd.read_sql("SELECT * FROM people", con=engine)
# people_df.head()
```

```python
salaries_df = pd.read_sql("SELECT * FROM salaries", con=engine)
# salaries_df.head()
```

```python
salaries_df = salaries_df.groupby('playerid').sum('salary')
```

```python
col_sal_df = pd.merge(college_df, salaries_df, how="inner", on='playerid')
```

```python
col_sal_peep_df = pd.merge(col_sal_df, people_df, how="inner", on='playerid')
```

```python
col_sal_peep_df
```

```python
col_sal_peep_df = col_sal_peep_df[['playerid', 'schoolid', 'salary', 'namefirst', 'namelast']]
```

```python
col_sal_peep_df
```

#### Vandy Only With Sort

```python
vandy = col_sal_peep_df[col_sal_peep_df['schoolid']=='vandy'].drop_duplicates()
vandy = vandy.sort_values('salary', ascending=False)
```
